```python
# Import dependencies
import pandas as pd
import json
```

```python
# Read JSON files into DataFrames
userProfileDataMediumArtists_df = pd.read_json('json_downloads/userProfileDataMediumArtists.json', lines=True)
userProfileDataLongArtists_df = pd.read_json('json_downloads/userProfileDataLongArtists.json', lines=True)
userProfileDataShortArtists_df = pd.read_json('json_downloads/userProfileDataShortArtists.json', lines=True)
userprofileTracks_df = pd.read_json('json_downloads/userProfileDataTracksApp.json', lines=True)

# Add a new column to each DataFrame to include the file name
userProfileDataMediumArtists_df = userProfileDataMediumArtists_df.assign(file_name='userProfileDataMediumArtists')
userProfileDataLongArtists_df = userProfileDataLongArtists_df.assign(file_name='userProfileDataLongArtists')
userProfileDataShortArtists_df = userProfileDataShortArtists_df.assign(file_name='userProfileDataShortArtists')
userprofileTracks_df = userprofileTracks_df.assign(file_name='userProfileDataTracksApp')

#combine User Profile dataframes into one DataFrame
combinedUserProfiles_df = pd.concat([userProfileDataMediumArtists_df, userProfileDataLongArtists_df,
                                    userProfileDataShortArtists_df, userprofileTracks_df ], ignore_index=True)

#export Combined User Profile dataframe to .csv so it can be used to populate the User Profile Table
combinedUserProfiles_df.to_csv('SQLite/csv/combinedUserProfile.csv', index=False)

print("DataFrame exported to combinedUserProfile.csv")
```

```python
#Read in Top Tracks JSON file into a Pandas DataFrame
topTracksLong_df = pd.read_json('json_downloads/topTracksArrayLong.json')
topTracksMedium_df = pd.read_json('json_downloads/topTracksArrayMedium.json')
topTracksShort_df = pd.read_json('json_downloads/topTracksArrayShort.json')
user_recentPlayArray_df = pd.read_json('json_downloads/recentlyPlayed.json')

#Merge all Top Tracks dfs into one Dataframe to populate "TopTracks" Table in SQL Database:
combinedTopTracks_df = pd.concat([topTracksLong_df, topTracksMedium_df, topTracksShort_df], ignore_index=True)

#Add in column with Top Tracks App name to connect it to "User Profile" table in SQL, 
#to ensure user data is identifiable by username
combinedTopTracks_df = combinedTopTracks_df.assign(file_name='userProfileDataTracksApp')

#export Top Tracks to .csv so it can be used to populate the Top Tracks Table
combinedTopTracks_df.to_csv('SQLite/csv/combinedTopTracks.csv', index=False)

print("DataFrame exported to combinedTopTracks.csv")
```

```python
# Read the Recently Played JSON file into a Pandas DataFrame
user_recentPlayArray_df = pd.read_json('json_downloads/recentlyPlayed.json')

#Add in column with Top Tracks App name to connect it to "User Profile" table in SQL, 
#to ensure user data is identifiable by username
user_recentPlayArray_df = user_recentPlayArray_df.assign(file_name='userProfileDataTracksApp')

# Export the Recently Played DataFrame to a CSV file so it can be used to set up the Recently PLayed Table
user_recentPlayArray_df.to_csv('SQLite/csv/recentPlayArray.csv', index=False)

print("DataFrame exported to recentPlayArray.csv")
```

```python
#Read in Top Artists JSON file into a Pandas DataFrame
user_topArtistsLong_df = pd.read_json('json_downloads/topArtistsLong.json')
#Add in column with Artist App name to connect it to "User Profile" table in SQL, 
#to ensure user data is identifiable by username
user_topArtistsLong_df = user_topArtistsLong_df.assign(file_name='userProfileDataLongArtists')

user_topArtistsMedium_df = pd.read_json('json_downloads/topArtistsMedium.json')
#Add in column with Artist App name to connect it to "User Profile" table in SQL, 
#to ensure user data is identifiable by username
user_topArtistsMedium_df = user_topArtistsMedium_df.assign(file_name='userProfileDataMediumArtists')

user_topArtistsShort_df = pd.read_json('json_downloads/topArtistsShort.json')
#Add in column with Artist App name to connect it to "User Profile" table in SQL, 
#to ensure user data is identifiable by username
user_topArtistsShort_df = user_topArtistsShort_df.assign(file_name='userProfileDataShortArtists')
 
#combine Top Artists into one DataFrame
combinedTopArtists_df = pd.concat([user_topArtistsLong_df, user_topArtistsMedium_df, user_topArtistsShort_df], ignore_index=True)

#export Top Artists to .csv so it can be used to populate the Top Artists Table
combinedTopArtists_df.to_csv('SQLite/csv/combinedTopArtists.csv', index=False)

print("DataFrame exported to combinedTopArtists.csv")
```

```python
# Path to the JSON file
file_path = 'json_downloads/relatedArtistsMedium.json'

# Read data from the JSON file
with open(file_path, 'r') as file:
    json_data = json.load(file)

# Extract the nested dictionaries and create a DataFrame
nested_dicts = []
for item in json_data:
    if 'relatedArtists' in item:
        nested_dicts.extend(item['relatedArtists'])

# Create a DataFrame from the nested dictionaries
relatedArtistsMedium_df = pd.DataFrame(nested_dicts)

#Add in column with Artist App name to connect it to "User Profile" table in SQL, 
#to ensure user data is identifiable by username
relatedArtistsMedium_df = relatedArtistsMedium_df.assign(file_name='userProfileDataMediumArtists')

relatedArtistsMedium_df.head()
```

```python
# Path to the JSON file
file_path = 'json_downloads/relatedArtistsLong.json'

# Read data from the JSON file
with open(file_path, 'r') as file:
    json_data = json.load(file)

# Extract the nested dictionaries and create a DataFrame
nested_dicts = []
for item in json_data:
    if 'relatedArtistsLong' in item:
        nested_dicts.extend(item['relatedArtistsLong'])

# Create a DataFrame from the nested dictionaries
relatedArtistsLong_df = pd.DataFrame(nested_dicts)

#Add in column with Artist App name to connect it to "User Profile" table in SQL, 
#to ensure user data is identifiable by username
relatedArtistsLong_df = relatedArtistsLong_df.assign(file_name='userProfileDataLongArtists')

relatedArtistsLong_df.head()
```

```python
#combine related artists into one DataFrame
combinedRelatedArtists_df = pd.concat([relatedArtistsLong_df, relatedArtistsMedium_df], ignore_index=True)

#drop three columns from final dataframe, as these columns are populated with inconsistent data 
#(and columns provide fourth image size, and three are already available for each item)
combinedRelatedArtists_df = combinedRelatedArtists_df.drop(columns=['tArtistimage4_height', 'tArtistimage4_url', 'tArtistimage4_width'])

#export Top Tracks to .csv so it can be used to populate the Top Tracks Table
combinedRelatedArtists_df.to_csv('SQLite/csv/combinedRelatedArtists.csv', index=False)

print("DataFrame exported to combinedRelatedArtists.csv")
```
