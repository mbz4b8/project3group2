```python
import sqlite3
import csv
import json
```

```python

# Connect to SQLite database
conn = sqlite3.connect('Spotify.db')
cursor = conn.cursor()

# Create a table (if it doesn't exist)
cursor.execute('''CREATE TABLE IF NOT EXISTS "combinedUserProfile" (
                    "displayName" VARCHAR NOT NULL,
                    "file_name" VARCHAR NOT NULL,
                    CONSTRAINT "pk_combinedUserProfile" PRIMARY KEY ("file_name")
                )''')

# Read data from CSV file and insert into the table
with open('csv/combinedUserProfile.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row if it exists
    for row in reader:
        cursor.execute('INSERT OR IGNORE INTO combinedUserProfile ("displayName", "file_name") VALUES (?, ?)', row)

# Commit changes and close connection
conn.commit()
conn.close()
```

```python
# Connect to SQLite database
conn = sqlite3.connect('Spotify.db')
cursor = conn.cursor()

# Query to retrieve all tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

# Fetch all tables
tables = cursor.fetchall()

# Initialize a dictionary to store table data
data_dict = {}

# Iterate over tables
for table in tables:
    table_name = table[0]
    # Query to fetch all data from the current table
    cursor.execute(f"SELECT * FROM {table_name};")
    # Fetch column names
    column_names = [description[0] for description in cursor.description]
    # Fetch all rows from the current table
    rows = cursor.fetchall()
    # Convert rows to a list of dictionaries
    table_data = [dict(zip(column_names, row)) for row in rows]
    # Store table data in the dictionary
    data_dict[table_name] = table_data

# Convert data dictionary to JSON format
json_data = json.dumps(data_dict, indent=4)

# Write JSON data to a file
with open('tables_data.json', 'w') as file:
    file.write(json_data)

# Close the database connection
conn.close()
```
